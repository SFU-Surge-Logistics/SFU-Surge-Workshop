# kNN Classification Demo
## Dataset: Iris (flower species classification), default beginner data from sklearn

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load
iris = load_iris(as_frame=True)
df = iris.frame.rename(columns=dict(zip(iris.feature_names,
    ["sepal_len","sepal_wid","petal_len","petal_wid"])))
df["species"] = df["target"].map(dict(enumerate(iris.target_names))) 
```

```python
# explore
print(df.head())
print(df.select_dtypes(include=[np.number]).describe().T[["mean","std","min","max"]])
print(df["species"].value_counts())
```

```python
# Histograms
df[["sepal_len","sepal_wid","petal_len","petal_wid"]].hist(bins=15, figsize=(8,6))
plt.suptitle("Feature Distributions"); plt.show()

# Scatter matrix
scatter_matrix(df[["sepal_len","sepal_wid","petal_len","petal_wid"]], figsize=(8,8), diagonal="hist")
plt.suptitle("Pairwise Relationships"); plt.show()
```

```python
# split into train/test sets

X = df[["sepal_len","sepal_wid","petal_len","petal_wid"]].values
y = iris.target

X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
# train model

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_tr, y_tr)

y_pred = knn.predict(X_te)
print("Accuracy:", accuracy_score(y_te, y_pred))
print(classification_report(y_te, y_pred, target_names=iris.target_names))
```
